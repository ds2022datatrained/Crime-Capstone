```python
# First we importing the required libraries for the Crime in India Project analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from scipy.stats import pearsonr, spearmanr
from scipy.stats import chi2_contingency
```

Phase-2, Req.No.-2.1, we are going for the Graphical & Non Graphical Bivariate Analysis between Literacy Rate vs Total Crimes here

```python
total_crime_df=pd.read_excel('totalcrime.xlsx')
```

```python
x=total_crime_df['Average Literacy Rate']
```

```python
y=total_crime_df['TOTAL IPC CRIMES']
```

```python
plt.scatter(x,y)
plt.xlabel('Literacy Rate')
plt.ylabel('Total Crimes')
plt.title('Literacy Rate Vs Total Crimes')
plt.show()
# With the below scatter plot, we found that literacy rate is having a little negative relationship with crime among states
# Even though it is not very strong negative relationship, but most of the crimes are there with the lower literacy rate.
# Up to 80% leteracy rate maximum crimes are there, but after above 80% of leteracy rate the crimes are found very minimum
# Even up to 80% literacy, most of the crimes are lying within the 70% of literacy rate.
```

```python
plt.plot(x,y)
plt.xlabel('Literacy Rate')
plt.ylabel('Total Crimes')
plt.title('Literacy Rate Vs Total Crimes')
plt.show()
# The same observations of scatter plot is also showing here with line plot
```

```python
plt.figure(figsize=(20,20))
sns.boxplot(x='Average Literacy Rate', y='TOTAL IPC CRIMES', data=total_crime_df)
plt.xlabel('Literacy Rate')
plt.ylabel('Total Crimes')
plt.title('Literacy Rate Vs Total Crimes')
plt.show()
# By the below box plot it is clearly showing that most of the crime data lying within 67% of literacy rate among States/UT
# Rest of the small-small crimes are there above 67% of literacy rate without any outliers
```

```python
plt.hexbin(x, y, gridsize=30, cmap='YlOrBr')
plt.colorbar()
plt.xlabel('Literacy Rate')
plt.ylabel('Total Crimes')
plt.title('Literacy Rate Vs Total Crimes')
plt.show()
# Below hexbin plot also showing that crime data are most spreaded within the lower portion of literacy rate
```

```python
sns.jointplot(x='Average Literacy Rate', y='TOTAL IPC CRIMES', data=total_crime_df, kind='scatter')
plt.show
# In below joint plot, most of the crime data points are lying left side of 80% literacy rate
```

```python
subset_df = total_crime_df[['Average Literacy Rate', 'TOTAL IPC CRIMES']]
correlation_matrix = subset_df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Literacy Rate Vs Total Crimes')
plt.show()
# The below heatmap is clearly showing the little negative corelation of literacy with crime but not very strong relation.
```

```python
plt.figure(figsize=(10,10))
plt.bar(total_crime_df['Average Literacy Rate'], total_crime_df['TOTAL IPC CRIMES'],color=['green', 'red'])
plt.xlabel('Literacy Rate')
plt.ylabel('Total Crimes')
plt.title('Literacy Rate Vs Total Crimes')
plt.show()
# In below bar plot the Green Bar is for Lieteracy Rate & Red Bar is for Crime Cases.
# So we are seeing that maximum crime case are there within 80% of literacy rate & above 80% very minimum crimes are there.
```

```python
plt.figure(figsize=(20,20))
sns.violinplot(x=total_crime_df['Average Literacy Rate'], y=total_crime_df['TOTAL IPC CRIMES'])
plt.xlabel('Literacy Rate')
plt.ylabel('Total Crimes')
plt.title('Literacy Rate Vs Total Crimes')
plt.show()
# Similarly of the box plot the violin plot also showing the most crime data are lying within 70% of literacy rate as below
```

```python
pearson_corr, pearson_p_value = pearsonr(total_crime_df['Average Literacy Rate'], total_crime_df['TOTAL IPC CRIMES'])
print(f"Pearson correlation: {pearson_corr}, p-value: {pearson_p_value}")
spearman_corr, spearman_p_value = spearmanr(total_crime_df['Average Literacy Rate'], total_crime_df['TOTAL IPC CRIMES'])
print(f"Spearman correlation: {spearman_corr}, p-value: {spearman_p_value}")

# In below Pearson Corelation result -0.34 is a weak negative relationship between literacy rate and crime cases.
# So by increase of literacy rate the crime tends to decrease but not very strongly, but it will support to decrease the crime
# As the Probability Value 0.046 is below the default significance level of 0.05, which shows that the corelation is not random
# Similarly Pearson results, The Spearman results are showing the same conclusion with more strongly.
```
